```python
import pandas as pd
```

```python
ref_df = pd.read_csv('data/COSMIC_v3.2_SBS_GRCh38.txt', sep='\t', index_col=0)
ref_df
```

```python
test_df = pd.read_csv('data/test.SBS96.all', sep='\t', index_col=0)
test_df = test_df / sum(test_df['sample'])
```

```python
sum(test_df['sample'])
```

```python
from numpy import dot
from numpy.linalg import norm
dot(ref_df['SBS1'], test_df['sample']) / (norm(ref_df['SBS1']) * norm(test_df['sample']))
```

```python
list_cos_sim = []
for col in ref_df.columns:
    #cos_sim = ref_df[col] @ test_df['sample'] / ((ref_df[col] @ ref_df[col]) ** 1/2 * (test_df['sample'] @ test_df['sample']) ** 1/2)
    cos_sim = dot(ref_df[col], test_df['sample']) / (norm(ref_df[col]) * norm(test_df['sample']))
    print(cos_sim * 100)
    list_cos_sim.append(cos_sim)
```

```python
df_cos_sim = pd.DataFrame(list_cos_sim, index=ref_df.columns, columns=['cosine similarity'])
df_cos_sim.sort_values(by='cosine similarity', ascending=False)
```
